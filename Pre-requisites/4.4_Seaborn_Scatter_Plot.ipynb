# SeaBorn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

### plotting Scatter graph

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
```

```python
# dividing the graph according to 'sex' column
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex')
```

```python
plt.figure(figsize = (16,9))
# other important parameters...
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'sex',
                style = 'who', size = 'who', #stylling and sizing the graph according to 'who' column
                sizes = (30,60),  # setting the scatters min and max size
                
                palette = 'magma'  # setting the colour map..
               )
```

```python
# retrieving number of alives
sns.scatterplot(x = 'who', y = 'fare', data = titanic_df, hue = 'alive',
                style = 'alive', size = 'who', #stylling and sizing the graph according to 'who' column
                sizes = (30,60))
```

```python
# Plotting Multiple Type of Graphs in a single plot..
plt.figure(figsize = (16,9))

sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
sns.lineplot(x = 'age', y = 'fare', data = titanic_df)
sns.barplot(x = 'age', y = 'fare', data = titanic_df)
```
