# Pandas interpolate()   Method

````python
import pandas as pd
````

#### interpolate() is such a effective method, used in filling 'NaN' values with Guessing..
##### Lets open a .csv file

````python
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv')
df
````

````python
# interpolate() guess only Numeric Values, it don't Guess String Values

df.interpolate()
````

## Use of method  Parameter

````python
# use of method parameter : default value is 'linear'

df.interpolate(method = 'linear')
````

### ```Interpolate with (method = 'time')```

````python
df.interpolate(method = 'time')
# this will only work on dateTimeIndex Dataframe
````

````python
type(df.Date[0]) # we can see that the Date column is written in String format.
````

````python
# Chaning datatype of 'Date' column from 'string' into 'DateTime Datatype'

df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv', parse_dates = ['Date'])
df
````

````python
# Now the 'Date' Column Has been Changed into DateTime DataType
type(df.Date[0])
````

````python
# Changing 'Date' column into an index..

df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv', parse_dates = ['Date'], index_col = 'Date')
df
````

````python
# Now we can interpolate with (method = 'time')..

df.interpolate(method = 'time')
# this will guess the values in the basis of pervious date index...
````

### ```Interpolate with (method = 'index')```

````python
# Changing 'Class' column into an index..

df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv', index_col = 'Class')
df
````

````python
# Now we can interpolate with (method = 'index')..

df.interpolate(method = 'index')

# this will guess the values in the basis of index...
#i.e. all columns with same index will be filled with same value..
````

### ```Interpolate with (method = 'nearest')```

````python
df.interpolate(method = 'nearest')
#fills on the basis of nearest upper values
````

### ```Interpolate with (method = 'polynomial')```

````python
# Changing 'Date' column into an index..

df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv', parse_dates = ['Date'], index_col = 'Date')
df
````

````python
df.interpolate(method = 'polynomial', order = 1) # with order 1 of polynomial
````

````python
df.interpolate(method = 'polynomial', order = 2) # with order 2 polynomial
````

````python
df.interpolate(method = 'spline', order = 2) # with order 2 spline
````

## Use of axis parameter

````python
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\student_10.csv')
df
````

````python
# two types  --> axis = 0  => fill colomn-by-colomn
#            --> axis = 1  => fill row-by-row

df.interpolate() # by default the axis is 0
````

````python
# to use the axis row-by-row, there should not be any object datatype in the Dataframe..
df.dtypes
````

````python
# ALL The DATATYPES SHOULD BE SAME TO INTERPOLATE ROW-BY-ROW

df.interpolate(axis = 1)  # so, it will show error..
````

## Use of  limit  &  limit_direction  Parameter

````python
# filling one NaN value in each column
df.interpolate(limit = 1)
````

````python
# filling two NaN value in each column
df.interpolate(limit = 2)
````

````python
# filling two NaN values in each column from backward direction
df.interpolate(limit = 2, limit_direction = 'backward')
````

````python
# filling two NaN values in one from forward and one from backward direction
df.interpolate(limit = 2, limit_direction = 'both')
````

## Use of limit_area Parameter

````python
# default value is 'inside'..

df.interpolate(limit_area = 'outside')
# no missing values have filled.. 
````

### Changing the actual dataframe without creating new one

````python
df.interpolate(inplace = True)
df
````
