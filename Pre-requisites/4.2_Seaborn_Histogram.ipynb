# SeaBorn Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

### Plotting Histogram from the above DataFrame..

```python
# sns.distplot() function is used to Plot Histogram...
sns.distplot(tips_df['size'])
plt.show()
```

```python
sns.distplot(tips_df['tip'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'])
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], bins = 100) # 'bins' parameter is used to set the range..
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], hist = False) # 'hist' is used to enable/disable the Histogram..
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], kde = False) # 'kde' is used to enable/disable the Line Plot..
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], rug = True) # 'rug' is used to enable/disable the Rug Plot on the Axis..
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], fit = norm) # Plotting The Normal of the Data in a Line..
plt.show()
```

```python
sns.distplot(tips_df['total_bill'], color = 'r') # 'kde' is used to change the colour..
plt.show()
```

### Other important parameters of sns.distplot()

```python
sns.set()

sns.distplot(tips_df['total_bill'], label = 'Total Bill')
plt.title("Histogram Of Total Bill")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
# Setting The Range... 
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set()

sns.distplot(tips_df['total_bill'], bins = bins)

plt.xticks(bins)
plt.title("Histogram of Total Bill")

plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set()

# Other Important Parameter of distplot() Method..
sns.distplot(tips_df['total_bill'],
             # Configuring Style of Histogram
            hist_kws = {'color':'#FFFF00', 'edgecolor':'#FF4500', 'linewidth':5, 'linestyle':'-'},
             
             # Configuring Style of Line Plot
            kde_kws = {'color' : 'b', 'linewidth':4, 'linestyle':'-', 'alpha':0.9},
            
             # Configuring Style of Rug-Plot
            rug = True,
            rug_kws = {'color':'g', 'linewidth':5, 'linestyle':'-.', 'alpha':0.8}
            label  'Total Bill')

plt.title("Histogram of Resturant Total Bill", fontsize = 20)
plt.title("Total Bill", fontsize = 15)
plt.legend()

plt.show()
```

```python
plt.figure(figsize = (16,9))

#Plotting Multiple Histograms..
sns.distplot(tips_df["total_bill"], bins = 9, label = "Total Bill")
sns.distplot(tips_df["tip"], bins = 9, label = "Tips")
sns.distplot(tips_df["size"], bins = 9, label = "Size")

plt.legend()
plt.show()
```
