# MatPlotLib Histogram

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))
````

````python
print(ml_students_age)
print(py_students_age)
````

### Plotting Histogram

````python
# plt.hist() is used to plot Histograms...
plt.hist(ml_students_age)

plt.title("ML Students Age Histogram")   # Setting the Title of Histogram
plt.xlabel("Students Age Category")  # Labelling X_Axis and Y_axis
plt.ylabel("No. Of Students")

plt.show()
````

````python
# Setting the range Manually...
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_students_age, bins)

plt.title("ML Students Age Histogram")   # Setting the Title of Histogram
plt.xlabel("Students Age Category")  #Labelling X_Axis and Y_axis
plt.ylabel("No. Of Students")

plt.show()
````

#### ```Discussing other Parameters of plt.hist() Method```

````python
# Setting the range Manually...
bins = [15, 20, 25, 30, 35, 40, 45]

# 'rwidth' is to set the width of 'bar', set using 'histtype' parameter..
# and the 'color' parameter is used to set the colour of bar..
# orientation is to set the bar's orientation i.e. vertical or horizontal...

plt.hist(ml_students_age, bins, rwidth = 0.8, orientation = 'horizontal', histtype = 'bar', color = 'r')

plt.title("ML Students Age Histogram")   # Setting the Title of Histogram
plt.xlabel("Students Age Category")  #Labelling X_Axis and Y_axis
plt.ylabel("No. Of Students")

plt.show()
````

### Plotting Multiple Graph in a Graph

````python
# Setting the range Manually...
bins = [15, 20, 25, 30, 35, 40, 45]

# two dataframes should be given in a list..

plt.figure(figsize = (16,9))
plt.hist([ml_students_age, py_students_age], bins, rwidth = 0.8, orientation = 'vertical', 
         histtype = 'bar', color = ['r','y'], label = ['ML Students','Py Student'])

plt.title("ML Students Age Histogram")   # Setting the Title of Histogram
plt.xlabel("Students Age Category")  #Labelling X_Axis and Y_axis
plt.ylabel("No. Of Students")
plt.legend() # Without this 'label' will not show..

plt.show()
````
