# Pandas Pivot Table
#### The Levels in the Pivot Table will be stored in Multi-Index Objects (Hierarchical Indexes) on the Index and Columns of the Result DataFrame

```python
import pandas as pd
```

```python
df= pd.read_csv('D:\\SAYAN SRK\\Datasets\\Movie_ratings.csv')
df
```

```python
#pivot_table() is used to aggregate specific rows and columns

# Row-wise Aggregation according to Genre
df.pivot_table(index = 'Genre')

# by default the table shows the Mean values..
```

```python
# Year-wise(Column-wise) Aggregation according to Genre(Row)..

df.pivot_table(index = 'Genre', columns = 'Year')
```

```python
# displaying Counts according to 'Count()'..

df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
```

```python
# displaying Counts according to 'sum()'..

df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum')
```

```python
# displaying Counts according to 'Max()'..

df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max')
```

```python
# There is some missing values in the output...We can set these values with given inputs.
# fill_value parameter is used to do this..

df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum', fill_value = 0)
```

```python
# Used to get the grand total.. for each columns..

df.pivot_table(index = 'Genre', columns = 'Year', fill_value = 0, margins = True)
```

```python
# Used to get the grand total.. for each columns..

df.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum', fill_value = 0, margins = True)
```

# Pandas melt() Method
#### Pandas melt() Function is used to Transform or Reshape Data..

```python
df= pd.read_csv('D:\\SAYAN SRK\\Datasets\\Movie_ratings.csv')
df
```

```python
# Getting the dataframe as Long Format..
pd.melt(df)
```

```python
# Getting an extra column, here it is 'Genre'..
pd.melt(df, id_vars = ['Genre'])
```

```python
# Getting an extra column, here it is 'Year'..
pd.melt(df, id_vars = ['Year'])
```

```python
# showing only specific variable in the melted Dataframe..

pd.melt(df, id_vars = ['Year'], value_vars = ['Genre'])
```

```python
pd.melt(df, id_vars = ['Genre'], value_vars = ['Rating%'])
```

```python
# Setting desired 'Varibale_Name' and 'Value_Name' in the Column Header..

pd.melt(df, id_vars = ['Genre'], value_vars = ['Rating%'], var_name = 'Category', value_name = 'Data')
```
