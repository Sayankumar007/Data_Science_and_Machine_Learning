# MatPlotLib Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\googleplaystore.csv', nrows = 5000)
df.shape
```

```python
# setting 'Rating' column as X axis and 'Reviews' column as Y axis 
x = df["Rating"]
y = df["Reviews"]
```

### Plotting Scatter Graph

```python
# plt.scatter() method is used to Plot the scatter graph..
plt.scatter(x,y)

# Setting Title and Labels for X and Y axis of the graph...
plt.title("Google Play Store Apps")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

### Other Parameters of plt.scatter() Method

```python
plt.figure(figsize = (16,9))

# Here are some other Parameters of plt.scatter() Method..
plt.scatter(x,y, c = 'r', marker = 'o', s = 50, alpha = 0.4, linewidths = 2, edgecolors = 'b')

# c => used to set colours
# marker => used to set the point style, such as circle, stars, triangles
# alpha => to set opacity
# linewidths => used to set the border width of each markers
# edgecolors => used to set the colour of the border.

# Setting Title and Labels for X and Y axis of the graph...
plt.title("Google Play Store Apps")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
```

### Setting Multiple Scatter in a Graph

```python
plt.figure(figsize = (16,9))

plt.scatter(x,y, c = 'r', marker = 'o', s = 25, alpha = 0.4, linewidths = 1, edgecolors = 'b', label = "Reviews")
plt.scatter(x, df['Installs'], c = 'y', marker = 'o', s = 25, alpha = 0.4, linewidths = 1, edgecolors = 'g',
            label = "Installs")

# Setting Title and Labels for X and Y axis of the graph...
plt.title("Google Play Store Apps")
plt.xlabel("Ratings")
plt.ylabel("Reviews and Installs")
plt.legend()

plt.show()
```
