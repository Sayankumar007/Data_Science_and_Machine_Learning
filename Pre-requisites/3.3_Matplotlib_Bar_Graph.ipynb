# MatPlotlib Bar Graph

````python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
````

````python
classes = ['C/C++','Python', 'DSA', 'Javacript', 'AI/ML']

class1_students = [30, 10, 20, 25, 10]
class2_students = [40, 5, 25, 20,10]
class3_students = [20, 35, 15, 5, 15]
````

### Plotting the Bar Graph..

````python
plt.bar(classes, class1_students)
plt.show()
````

### Important Parameters of plt.bar()

```x ```: float or array-like
    The x coordinates of the bars. See also *align* for the
    alignment of the bars to the coordinates.

```height``` : float or array-like
    The height(s) of the bars.

```width``` : float or array-like, default: 0.8
    The width(s) of the bars.

```bottom``` : float or array-like, default: 0
    The y coordinate(s) of the bars bases.

```align``` : {'center', 'edge'}, 
       default: 'center'
       Alignment of the bars to the *x* coordinates:

    - 'center': Center the base on the *x* positions.
    - 'edge': Align the left edges of the bars with the *x* positions.

    To align the bars on the right edge pass a negative *width* and
    ``align='edge'``.
    
```color``` : color or list of color, optional
    The colors of the bar faces.

```edgecolor``` : color or list of color, optional
    The colors of the bar edges.

```linewidth``` : float or array-like, optional
    Width of the bar edge(s). If 0, don't draw edges.
    
```alpha``` : changing the opacity of the colour.

```linestyle``` : sets the style of the edge.

````python
plt.bar(classes, class1_students, width = 0.7, align = 'edge', color = 'c',
        edgecolor = 'k', linewidth = 2, linestyle = '-')
plt.title("Students In Different Domains")
plt.xlabel("Domain Name")
plt.ylabel("No. of Students")

plt.show()
````

### Plotting Multiple Bar Graphs

````python
# making indexes for each domain classes..
classes_index = np.arange(len(classes))

width = 0.2

# plt.figure(figsize = (16,9))

# plotting graph according to positions...
plt.bar(classes_index, class1_students, width, color = 'r', label = 'Class.1_Students') # first 20% or 0.2 part of width

# positions will be  increased by 'width' to avoid overlapping
plt.bar(classes_index + width, class2_students, width, color = 'g', label = 'Class.2_Students') # next 40% width
plt.bar(classes_index + (2*width), class3_students, width, color = 'b', label = 'Class.3_Students') # next 60% width

# after the plotting we should replace the index values with its Proper Class Name..
# plt.xticks() method will do this..
plt.xticks(classes_index + (2*width), classes, rotation = 20) # rotation will occur by 20 degree..
plt.xticks(classes_index + width, classes, rotation = 20) # rotation will occur by 20 degree..

plt.title("Students In Different Domains", fontsize = 30)
plt.xlabel("Domain Name", fontsize = 20)
plt.ylabel("No. of Students", fontsize = 20)
plt.legend() # Necessary for showing Labels..

plt.show()
````
