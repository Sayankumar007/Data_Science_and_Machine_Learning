### Creating Complex Heat Map

```python
gw_df = pd.read_csv('d:\\Sayan SRK\\Datasets\\global_warming.csv')
gw_df
```

```python
# as heatmaps only allow numerical value so we have to remove numerical values

gw_df = gw_df.drop(columns = ['Country Code', 'Indicator Name', 'Indicator Code'], axis = 1).set_index('Country Name')
gw_df
```

```python
plt.figure(figsize = (16,9))
# Creating a complex Heatmap..
sns.heatmap(gw_df)
```

```python
plt.figure(figsize = (16,9))

# Setting Range in the Colour Bar..
sns.heatmap(gw_df, vmin = -1, vmax = 21)
```

```python
plt.figure(figsize = (16,9))

# Setting the Colour Map..
sns.heatmap(gw_df, cmap = "coolwarm")
```

```python
plt.figure(figsize = (16,9))

# Enable/Disable Annotations..
sns.heatmap(gw_df, annot = True)
```

### Other Important Parameters..

```python
plt.figure(figsize = (16,9))

# Parameters for annotation...
annot_kws = {'fontsize': 10,
            'fontstyle': 'italic',
            'color': 'k',
            'alpha': 0.7,
            'rotation':'vertical', # Sets the rotation of the annotation..
            'verticalalignment':'center'} # vertical positions.

# Applying the annotations..
sns.heatmap(gw_df, cmap = 'coolwarm', annot = True, annot_kws = annot_kws)
```

```python
plt.figure(figsize = (16,9))

# stylling the colour bar..
cbar_kws = {'orientation': 'horizontal',  # position : Horizontal or Vertical..
            'shrink': 1,  # size of coloubar
            
            'extend': 'min',      # 'min'-> min side will sharpen, 'max'-> max size will sharpen..
            'extendfrac': 0.1,    # Measure of extend
            
            'ticks' : np.arange(0,22),
            'drawedges': True
           }

sns.heatmap(gw_df, cmap = 'hot', 
            annot = True, annot_kws = annot_kws, # used to set annotations
            
            linewidth = 4, linecolor = 'w', # use to style each segment of heatmap...
            
            cbar = True,    # Enable/disable Colour Bar
            cbar_kws = cbar_kws,  # Stylling in Colour Bar
            
            xticklabels = True, yticklabels = True,   # Enable/disable Labels and values of the axes..
            
            
           )
```

### Stylling in labels

```python
plt.figure(figsize = (16,9))

ax = sns.heatmap(gw_df)

#setting title and axes labels..
ax.set(title = 'Heat Map for "Who is Responsible for Global Warming"',
       xlabel = "CO2 Emissions (metric ton per capital per year)",
       ylabel = "Country Name")

#setting font size..
sns.set(font_scale = 0.5) #scalling the previous font size..
```

## Correlation Heat Map

```python
# first we need the correlation of the matrix

gw_df_cr = gw_df.corr()  # .corr() method is used to create correlation graph
gw_df_cr
```

```python
plt.figure(figsize = (16,9))

# Plotting Correlation Heat Map
ax = sns.heatmap(gw_df_cr, annot = True, linewidth = 3, annot_kws = {'fontsize': 12})

# stylling the labels
ax.tick_params(size = 20, color = 'k', labelsize = 10, labelcolor = 'k')

plt.title('Heat Map for "Who is Responsible for Global Warming"', fontsize = 25)
plt.xlabel("CO2 Emissions (metric ton per capital per year)", fontsize = 20)
plt.ylabel("Country Name", fontsize = 20)

# 'method' parameter can also be use to create correlation Heat Map
```

## Feature Engineering Touch

```python
# lets import Large data set..
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()

cancer_dataset
```

```python
# creating dataFrame from the DataSet
cancer_df = pd.DataFrame(np.c_[cancer_dataset['data'], cancer_dataset['target']],
                         columns = np.append(cancer_dataset['feature_names'], ['target']))

cancer_df
```

```python
plt.figure(figsize = (25, 20))

ax = sns.heatmap(cancer_df.corr(), annot = True, linewidth = 3, annot_kws = {'fontsize': 12})

ax.tick_params(size = 20, color = 'k', labelsize = 10, labelcolor = 'k')

plt.title('Heat Map for "Breast Cancer Patients"', fontsize = 25)
```
