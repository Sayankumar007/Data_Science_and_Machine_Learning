## Pandas merge() Function
### ```Pandas merge() connects columns or indexes in DataFrame based on one or more keys```

````python
# Creating 1st DataFrame.. 
df1 = pd.DataFrame({ 'ID' : [1, 2, 3, 4],
                     'Class' : [9, 10, 11, 12]})
df1
````

````python
# Creating 2nd DataFrame.. 
df2 = pd.DataFrame({ 'ID' : [1, 2, 3, 5],
                   'Name' : ['A', 'B', 'C', 'D']})
df2
````

````python
# 'on' parameter is used to merge on the basis of a specific column

pd.merge(df1, df2, on = 'ID')  
# 'Index_Value' and 'Column_Name' given in the 'on' parameter should be common in both DataFrame
# output will ignore 4th index, as ID is diffrent
````

````python
# Changing Position.. 
pd.merge(df2, df1, on = 'ID')  
````

````python
# how parameter...is used to set priority..

pd.merge(df1, df2, on = 'ID', how = 'inner')  # 'inner' means intersection
````

````python
pd.merge(df1, df2, on = 'ID', how = 'left')  # 'left' means priority to left DataFrame

# see the Value of 'ID' in the 4th index
````

````python
pd.merge(df1, df2, on = 'ID', how = 'right')  # 'right' means priority to Right DataFrame

# see the Value of 'ID' in the 4th index
````

````python
pd.merge(df1, df2, on = 'ID', how = 'outer')  # 'outer' means no priority to any DataFrame

# see the Value of 'ID' in the 4th and 5th index, both are considered
````

````python
# indicator parameter is used to get the information of each row in the merged_dataframe

pd.merge(df1, df2, on = 'ID', how = 'outer', indicator = True)  
````

#### ```# What if.. both DataFrame has different ID value..```

````python
# Updating 2nd DataFrame.. 
df2 = pd.DataFrame({ 'ID' : [5, 6, 7, 8],
                   'Name' : ['A', 'B', 'C', 'D']})
df2
````

````python
pd.merge(df1, df2)

# this will not show any data.. as there is no common value in the ID column
````

````python
# to merge the columns with different value, we need 'left_index' and 'right_index' parameter

pd.merge(df1, df2, left_index = True, right_index = True)
````

## Pandas concat() Function
### ```Used for easily combining together Series, DataFrame and Panel Objects```

#### Concatinating Two Series

````python
# Creating 1st Series
sr1 = pd.Series([0,1,2])
sr1
````

````python
# Creating 2nd Series
sr2 = pd.Series([3,4,5,6,7])
sr2
````

````python
# Concatinating two Series..
pd.concat([sr1,sr2])
````

#### Concatinating Two DataFrames

````python
# Creating 1st DataFrame.. 
df1 = pd.DataFrame({ 'ID' : [1, 2, 3, 4],
                   'Name' : ['A', 'B', 'C', 'D'],
                  'Class' : [5, 6, 7, 8]})
df1
````

````python
# Creating 2nd DataFrame.. 
df2 = pd.DataFrame({ 'ID' : [5, 6, 7, 8],
                   'Name' : ['E', 'F', 'G', 'H'],
                  'Class' : [9, 10, 11, 12]})
df2
````

````python
# concatinating df1 and df2 Row-wise i.e. axis = 0
pd.concat([df1,df2])
````

````python
# concatinating df1 and df2 Column-wise i.e. axis = 1
pd.concat([df1,df2], axis = 1)
````

````python
# Use of ignore_index parameter :

# --> See previous example that indexes are not Systematic i.e. [0 1 2 3 0 1 2 3]
# --> Used to concatinate with Systematic Index i.e. [0 1 2 3 4 5 6 7]

pd.concat([df1,df2], axis = 0, ignore_index = True)
````

````python
# Keys parameter.. used to separate dataframes from each other...

pd.concat([df1, df2], keys = ['First DF', 'Second DF'])
````

````python
# Column-wise Seperation...

pd.concat([df1, df2], axis = 1, keys = ['First DF', 'Second DF'])
````

````python
# Concatinating DataFrame with No Common Columns..

# Creating 3rd DataFrame.. 
df3 = pd.DataFrame({ 'Marks' : [45, 56, 97, 68]})
df3
````

````python
pd.concat([df1, df3])  # Concatinating df1, df3
````

## Pandas join() Method
### ```join() is a Convenient Method for Combining```
### ```Two Potentially Differently-indexed Column```

````python
df1 = pd.DataFrame({'A' : [1,2,3],
                    'B' : [10,20,30]})

df2 = pd.DataFrame({'C' : [4,5,6],
                    'D' : [40,50,60]})

display(df1, df2)
````

````python
# Joining two DataFrames..

df1.join(df2)
````

````python
# Joning two unequal DataFrames..
df3 = pd.DataFrame({'C' : [4,5],
                    'D' : [40,50]})
df3
````

````python
# Printing Left Index-wise
df1.join(df3)
````

````python
# Printing Right Index-wise
df1.join(df3, how = 'right')
````

````python
# What if.. there is two Columns with same Name..

df4 = pd.DataFrame({'A' : [4,5,6],
                    'D' : [40,50,60]})
df4
````

````python
# If there is Two Columns with same Name, Join() method will show Error
df1.join(df4)
````

````python
# To avoid this error, we have to use 'lsuffix' and 'rsuffix' parameter..

df1.join(df4, lsuffix = '_1', rsuffix = '_2')

# we can also use only one of them..
````

## Pandas append() Method
### ```append() is used to append rows of other DataFrame to the```
### ```End of given DataFrame, returning New DataFrame Object..```

````python
df1 = pd.DataFrame({'A' : [1,2,3],
                    'B' : [10,20,30]})

df2 = pd.DataFrame({'A' : [4,5,6],
                    'B' : [40,50,60]})

display(df1, df2)
````

````python
df1.append(df2)

# however index will not be Systematic...
````

````python
df1.append(df2, ignore_index = True)

# Make the Indexes Systematic...
````
