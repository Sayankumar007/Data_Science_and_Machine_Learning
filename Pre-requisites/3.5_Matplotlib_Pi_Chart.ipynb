# MatPlotLib Pi-Chart 

```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'R', 'ML', 'AI', 'Data Science']
class1_students = [45, 15, 35, 25, 30]
```

### Plotting Pi_Chart

```python
plt.pie(class1_students, labels = classes)
plt.show()
```

```python
# Cutting out a piece from the Pi_Chart...
explode = [0, 0, 0.2, 0, 0] # this will cutout the plot of 3rd index i.e. ML by 0.2 unit

plt.pie(class1_students, labels = classes, explode = explode)
plt.show()
```

### Other parameters of plt.pie() Method

```python
color_lst = ['y', 'r', 'b', 'm', 'c']
explodes = [0, 0, 0.2, 0, 0.1]
wedgeprop_value = {'linewidth' : 4, 'width' : 1.3, 'edgecolor' : 'k'}  # 'width' is used to set a hole in the center..

# Important Parameters of plt.pie() Method..
plt.pie(
    class1_students,
    explode = explodes, # ***to cut out the important section of Pi-Chart (default => None)
    labels = classes,   # ***to set the Labels of each portion of Pi-Chart (default => None)
    
    colors = color_lst, # ***To set the colours of the portion.. (default => None)
    
    autopct = '%0.2f%%', # ***to set the auto-Percentage Floating value format (default => None)
    
    pctdistance=0.6,  # Distance of percentage value from the center..
    
    shadow = True, # to set the shadow for the pi-chart (default => False)
    
    labeldistance = 1.2,    # Distance of the labels from the Pi-Chart..
    
    startangle = 270, # Starting index will start at what angel? (default => 0)
    
    radius=1.5,  # Radius of the Circle (default => 1.1)
    
    counterclock=True,
    
    wedgeprops = wedgeprop_value,     # ***USED To SET THE EDGE OF THE PI_CHART..(default = None)
    
    textprops=None,
    center=(1, 2), # Setting the Center Co-ordinate (default => (0, 0))
    frame=False,
    rotatelabels=20, # Rotating Labels by 20 units..(default => False)
    normalize=True,
    data=None,
)

plt.show()
```
