### Accessing Elements from the DataFrame

#### ```-> Accessing Specific no. of Rows```

````python
# Accessing first Row..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', nrows = 1)
df
````

````python
# Accessing first five rows..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', nrows = 5)
df
````

#### ```-> Accessing columns with specific index..```

````python
# Accessing first column..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', usecols = [0]) 
# index should be given in square brackets..
df
````

````python
# Accessing 2nd, 3rd and 5th columns only..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', usecols = [1,2,4]) 
# index should be given in square brackets..
df
````

### Skipping specific no. of rows we don't need..

#### ```-> we have to give an integer, indicating how many rows to skip, in the parameter..```

````python
# Skipping first row only..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', skiprows = 1) 
df  #first row will skipped and the next will be treated as label row
````

````python
# Skipping first three rows only..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', skiprows = 3) 
df
````

### Skipping Rows with Specific Index..

#### ```-> we have to give the index in a square bracket, in the parameter```

````python
# skipping 3rd row (index = 2)
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', skiprows = [2]) 
df
````

````python
# skipping 3rd, 5th and 7th row (index = 2, 4, 6)
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', skiprows = [2,4,6]) 
df
````

### Using A Specific Column as The Index of Our DataFrame

#### ```SYNTAX : index_col = 'Column_name' or [Index]```

````python
# Making the Name as Index..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', index_col = 'Name') 
df
````

````python
# Using the ID as Index..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', index_col = 0) 
df
````

````python
# Using the Industry as Index..
df = pd.read_csv('D:\\SAYAN SRK\\Datasets\\fortune_10.csv', index_col = 2 ) 
df
````
