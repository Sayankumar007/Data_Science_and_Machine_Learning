#### As it is clear from the above dataset that the it contains many NULL valus or NaN values..So, we have to see how many NaN values it contains, and what is the percentage of NaN values in each column,... And We are gonna drop those columns with larger percentage of missing values..

```python
# Lets see how many numbers of missing values, Each column of the dataset contains.. 
df.isnull().sum()
```

### Lets visualize the current condition of the dataset..

```python
# Data Visualization is needed to know the dirtyness of the current dataset..
plt.figure(figsize = (25,25))
sns.heatmap(df.isnull())
```

### Lets see the percentage of missing values in Each column of the dataset.. 

```python
# Lets see the percentage of missing values in Each column of the dataset.. 
prcnt_null = df.isnull().sum()/df.shape[0]*100
```

### Here We Have To Get List Of The Column With 17% or More NaN values

```python
# Creating a list of columns with 17% or more NaN values, fro the 'prcnt_null' dictionary
dropped_columns = prcnt_null[prcnt_null > 17].keys()  # we need only column name so, consider only key values..
dropped_columns
```

### Dropping Columns with More Missing Values..

```python
df2_drp_clm = df.drop(columns = dropped_columns)
```

```python
# Now if we check the size, we can see that, 6 columns have been dropped from the dataset of 81 columns
# i.e. there will be only 75 columns in the datasets..

df2_drp_clm.shape
```

```python
# Now If We see the The HeatMap, it will look more cleaned...
plt.figure(figsize = (25,25))
sns.heatmap(df2_drp_clm.isnull())
```
